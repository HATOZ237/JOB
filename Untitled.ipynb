```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import os
import pickle
```

```python
names = ['Breast_cancer', 'Opt_digits', 'Iris', "Wine"]
filename = "RAND-TAB-SVC"
infile = open(filename,'rb')
new_dict = pickle.load(infile)
infile.close()
```

```python
for i in range(len(names)):
    liste = [0 for n in range(10)]
    plt.subplot(2, 2, i+1)
    for k in range(10):
        liste[k]  = sorted(new_dict[names[i]][k]["mean_test_score"])
    liste = np.array(liste)
    mean_train = np.mean(liste, axis=0)
    std_train = np.std(liste, axis=0)
    plt.plot(mean_train)
    plt.xscale("log")
    
                        
```

```python
def sorted(liste):
    first = liste[0]
    for i in range(len(liste)):
        if first < liste[i]:
            first = liste[i]
        if first >= liste[i]:
            liste[i] = first
    return liste
```
